Explanation on each column names

age: age of people in the population

sex: sex of person(1-Male , 0-Feamle)

cp: chest pain type ( 0-typical angina,1-atypical angina,2-non_anginal pain,3-asymptomatic)

trestbps: resting blood pressure

chol:serum cholesterol in mg/dl

fbs:fasting blood sugar > 120 mg/dl (1-True , 0-False)

restecg:resting electrocardiographic results (values 0-normal, 1-having ST-T wave abnormality,2-probable or definite left ventricular hypertrophy by Estes' criteria)

thalach:maximum heart rate achieved

exang: exercise induced angina(1-Presence, 0-Absence)

oldpeak: ST depression induced by exercise relative to rest

slope:the slope of the peak exercise ST segment(0-downslopping, 1-flat, 2-upslopping)

ca:number of major vessels (0-3) colored by flouroscopy

thal:1 - normal, 2- fixed defect , 3 - reversible defect, 0- Null

target: Heart disease(0-No , 1-Yes)

# Data Visualisation

```python
df_1=df.copy()
```

```python
df_1
```

```python
#Changing numerical values into strings for analysis
def chng(x):
    if x==1:
        return "YES"
    else:
        return "No"
df_1['target']=df_1['target'].apply(chng)
df_1
```

```python
#Pie chart for finding total percentage of the population affected
df_1['target'].value_counts().plot(kind='pie',autopct='%1.1f%%',startangle=90)
plt.axis('equal')
plt.title("Distribution of number of people affected ")
plt.show()
```

    We can see 51.3% of the overall population is affected

```python
df_1['sex']=df_1['sex'].apply(lambda x:"Male" if x==1 else "Female")
df_1
```

```python
#countplot for different age 

plt.figure(figsize=(15,7))
sns.countplot(x=df_1['age'],data=df_1)

plt.title("Age distribution in the Population")
plt.show()
```

```python
# Forming different age categories

def category(x):
    if x<40:
        return "Young Age"
    elif x>=40 and x<55:
        return "Middle Age"
    else :
        return "Elderly"
df_1['age_category']=df_1['age'].apply(category)
df_1
```

```python
#Using countplot finding population distribution in different age categories

ax=sns.countplot(x=df_1['age_category'],hue=df_1['sex'],palette="PuBuGn",order=['Young Age','Middle Age','Elderly'])
plt.title("Population distribution in different Age Categories",fontsize=9)
for bars in ax.containers:
    ax.bar_label(bars)
```

In all age categories males are more than females.

```python
# countplot for people effected by heart disease in each age category

vx=sns.countplot(x=df_1['target'],hue=df_1['age_category'],order=("YES","No"),color="Green",saturation=10)
for bars in vx.containers:
    vx.bar_label(bars)
plt.title("Heart Disease affected in each age category",fontsize=8)
plt.xlabel("Diagonized or not")
plt.show()
```

 More people in middle age has been diagonized with heart disease.

```python
# swarmplot for finding effected population in each group along with gender

sns.swarmplot(data=df_1,x='target',y='age',hue='sex',order=['YES','No'],marker="*",size=4)
plt.title("Affected indviduals in different ages and their sex")
plt.xlabel("Affected")
plt.show()
```

In young age males are highly affected and we can see that above 70 heart disease are seen more in females.

```python
def cpname(x):
    if x==0:
        return "typical angina"
    elif x==1:
        return "atypical angina"
    elif x==2:
        return "non_anginal pain"
    else:
        return "asymptomatic"
df_1['cp_name']=df_1['chest_pain_type'].apply(cpname)
```

```python
df_1
```

```python
# countplot for effected people with different chest pain

sns.countplot(data=df_1,x='target',hue='cp_name',order=['YES','No'])
plt.xlabel("Heart Disease")
plt.title("Chest pain type and Heart Disease")
plt.show()
```

From this we can understand that people with typical angina is less likely to be diagonized with heart disease. And people 
with non anginal pain is more likely to be affected with heart disease.

```python
#count plot for chest pain type in each gender

sns.countplot(data=df_1,x='sex',hue='cp_name',palette="pastel")
plt.xlabel("Gender")
plt.title("Chest pain type vs Gender")
plt.show()
```

```python
#violinplot for finding cholesterol level in the population basis of diagonized or not

sns.violinplot(data=df_1,x='target',y='chol',palette="coolwarm",errorbar=None,order=["YES","No"])
plt.title("Cholesterol vs Heart Disease")
plt.xlabel("Heart Disease")
plt.ylabel("Cholosterol")
plt.show()
```

people with cholestrol level around 200 and 300 is prone to heart disease. As increased cholesterol level can also cause heart disease.

```python
#violinplot for finding cholesterol level in the population basis of gender
sns.violinplot(data=df_1,x='sex',y='chol',palette="coolwarm",errorbar=None)
plt.title("Cholesterol vs Gender")
plt.xlabel("Gender")
plt.ylabel("Cholosterol")
plt.show()
```

 Average cholesterol is more for females than males.

```python
#barplot for finding blood pressue in the population basis of gender

sns.barplot(data=df_1,x='sex',y='resting_blood_pressure',errorbar=None,palette="colorblind")
plt.title("Blood Pressure vs Gender")
plt.xlabel("Gender")
plt.show()
```

Average blood pressure is almost same in male and female

```python
#barplot for finding blood pressue in the population basis of heart disease
sns.boxplot(data=df_1,x='target',y='resting_blood_pressure',order=["YES","No"],palette="colorblind")
plt.title("Blood Pressure vs Heart Disease")
plt.xlabel("affected or not")
plt.show()
```

 This shows that blood pressure is not related to the heart disease.

```python
df_1['exang']=df_1['exercise_induced_angina'].apply(lambda x:"Presence" if x==1 else "Absence")
df_1
```

```python
#countplot for finding exercise induced angina in population

sns.countplot(data=df_1,x='target',hue='exang',palette="husl",order=["YES","No"],hue_order=["Presence","Absence"])
plt.xlabel("Heart Disease")
plt.title("Exercise induced angina and Heart Disease")
plt.show()
```

This shows that people having exercise induced angina which is caused from stress or over exercise doesnt probably have
heart disease.

```python
df_1=df_1.rename(columns={"fasting_blood_pressure":"fasting_blood_sugar"})
```

```python
df_1["fasting_blood_sugar"]=df_1["fasting_blood_sugar"].apply(lambda x:">120" if x==1 else "<120")
df_1
```

```python
# countplot for blood sugar level and heart disease

sns.countplot(data=df_1,x='fasting_blood_sugar',hue="target",palette="cubehelix",hue_order=["YES","No"],order=[">120","<120"])
plt.xlabel("Fasting Blood Sugar Level")
plt.legend(loc="upper left")
plt.title("Fasting Blood Sugar Level and Heart Disease")
plt.show()
```

This countplot shows that a high blood sugar doesnot always lead to heart disease

```python
# countplot for blood sugar level and gender

sns.countplot(data=df_1,hue='sex',x="fasting_blood_sugar",palette="deep",order=[">120","<120"])
plt.xlabel("Fasting Blood Sugar Level")
plt.legend(loc="upper left")
plt.title("Fasting Blood Sugar Level vs Gender")
plt.show()
```

High blood sugar level is seen in male.

```python
#A lineplot for bloodpressure level in different ages

plt.figure(figsize=(9,3))
sns.lineplot(x="age",y="resting_blood_pressure",hue="age_category",data=df_1)
plt.title("Blood Pressure vs age")
plt.show()
```

Here we can see blood pressure increases at elder age ie from 55

```python
#A lineplot for cholesterol level in different ages

plt.figure(figsize=(9,3))
sns.lineplot(x="age",y="chol",hue="age_category",data=df_1)
plt.title("Cholesterol vs age")
plt.show()
```

Cholesterol tends to gain a peak at age 55 and maintain that along elder age

```python
#A lineplot for thalach level in different ages

plt.figure(figsize=(9,3))
sns.lineplot(x="age",y="thalach",hue="age_category",data=df_1)
plt.title("Thalach vs age")
plt.show()
```

Maximum heart rate is achieved at young ages and eventually reduces during middle and elder ages.

```python
#Histplot for showing oldpeak values and heart disease

sns.histplot(x="oldpeak",data=df_1,bins=5,color="y",multiple="stack",hue="target",kde=True)
plt.show()
```

old peak value between 0-1.2 shows more chances for heart disease.
